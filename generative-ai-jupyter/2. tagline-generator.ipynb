```python
from vertexai.preview.generative_models import GenerativeModel

model = GenerativeModel("gemini-1.0-pro")

prompt = """
Cymbal Direct is partnering with an outdoor gear retailer. They\'re launching a new line of products designed to encourage young people to explore the outdoors. Help them create catchy taglines for this product line.

input: Help me to write a tagline about Cynic Direct outdoor durable, and lightweight backpack product, entertain the young hikers, family, enthusiast and so on
output: Cynic Direct: Better than any outdoor backpack

input: Empower the customer to willingly buy the product and try to make them believe that our product is really great
output: The ultimate backpack for any activities

input: integrates with natures like forest, or mountain please
output:
"""

responses = model.generate_content(
    prompt,
    generation_config={
        "temperature": 0.5,
        "max_output_tokens": 2048,
        "top_p": 1.0,
        "top_k": 40,
    },
    stream=True
    )

for response in responses:
    print(response.text)
```
